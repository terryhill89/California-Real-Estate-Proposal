```python
# Import Dependencies
import pandas as pd
import numpy as np
```

# Save data to dataframes

```python
# Save 2021 student data to dataframe
all_students_2021_df = pd.read_csv("Education_Resources/sb_ca2021_all_csv_v2.txt", sep="^", low_memory=False)
all_students_2021_df
```

```python
# Save 2021 entities data to dataframe
entities_2021_df = pd.read_csv("Education_Resources/sb_ca2021entities_csv.txt", sep="^", encoding='latin1')
entities_2021_df
```

```python
# Save 2022 students data to Dataframe
all_students_2022_df = pd.read_csv("Education_Resources/sb_ca2022_all_csv_v1.txt", sep="^")
all_students_2022_df
```

```python
# save 2022 entities data to dataframe
entities_2022_df = pd.read_csv("Education_Resources/sb_ca2022entities_csv.txt", sep="^", encoding="latin1")
entities_2022_df
```

# Clean all_students_2021_df

```python
# Merge entities and students dataframes
merged_2021 = pd.merge(all_students_2021_df, entities_2021_df, on=("County Code", "District Code", "School Code", "Test Year", "Type ID"), how="left")
merged_2021.info()
```

```python
# Drop Filler columns
merged_2021 = merged_2021.drop("Filler_x", axis=1)
merged_2021 = merged_2021.drop("Filler_y", axis=1)
```

```python
# Drop extra columns
merged_2021 = merged_2021.drop("Total Tested at Reporting Level", axis=1)
merged_2021 = merged_2021.drop("Total Tested with Scores at Reporting Level", axis=1)
merged_2021 = merged_2021.drop("Mean Scale Score", axis=1)
merged_2021 = merged_2021.drop("Area 1 Percentage Above Standard", axis=1)
merged_2021 = merged_2021.drop("Area 1 Percentage Near Standard", axis=1)
merged_2021 = merged_2021.drop("Area 1 Percentage Below Standard", axis=1)
merged_2021 = merged_2021.drop("Area 2 Percentage Above Standard", axis=1)
merged_2021 = merged_2021.drop("Area 2 Percentage Near Standard", axis=1)
merged_2021 = merged_2021.drop("Area 2 Percentage Below Standard", axis=1)
merged_2021 = merged_2021.drop("Area 3 Percentage Above Standard", axis=1)
merged_2021 = merged_2021.drop("Area 3 Percentage Near Standard", axis=1)
merged_2021 = merged_2021.drop("Area 3 Percentage Below Standard", axis=1)
merged_2021 = merged_2021.drop("Area 4 Percentage Above Standard", axis=1)
merged_2021 = merged_2021.drop("Area 4 Percentage Near Standard", axis=1)
merged_2021 = merged_2021.drop("Area 4 Percentage Below Standard", axis=1)
```

```python
# Check that columns have been dropped
merged_2021.info()
```

```python
merged_2021
```

```python
# Drop rows with missing data
merged_2021 = merged_2021[merged_2021["Students Tested"] != "0"]
merged_2021 = merged_2021[merged_2021["Students Tested"] != "*"]
merged_2021 = merged_2021[merged_2021["Percentage Standard Met"] != "*"]
merged_2021
```

```python
# Drop rows with california, County, or District as the Type ID (4, 5, 6,)
merged_2021 = merged_2021[merged_2021["Type ID"] != 4]
merged_2021 = merged_2021[merged_2021["Type ID"] != 5]
merged_2021 = merged_2021[merged_2021["Type ID"] != 6]
merged_2021
```

```python
# Check df again for null values
merged_2021.count()
```

```python
# Change Data Types
merged_2021 = merged_2021.astype({"Students Enrolled": int}, errors="raise")
merged_2021 = merged_2021.astype({"Students Tested": int}, errors="raise")
merged_2021 = merged_2021.astype({"Percentage Standard Exceeded": float}, errors="raise")
merged_2021 = merged_2021.astype({"Percentage Standard Met": float}, errors="raise")
merged_2021 = merged_2021.astype({"Percentage Standard Met and Above": float}, errors="raise")
merged_2021 = merged_2021.astype({"Percentage Standard Nearly Met": float}, errors="raise")
merged_2021 = merged_2021.astype({"Percentage Standard Not Met": float}, errors="raise")
merged_2021 = merged_2021.astype({"Students with Scores": int}, errors="raise")
merged_2021 = merged_2021.astype({"Zip Code": int}, errors="raise")
```

```python
# Double check column data types
merged_2021.info()
```

```python
# Drop the one row with missing data in Percentages
merged_2021 = merged_2021.dropna(how="any")
merged_2021.info()
```

```python
# Count number of zip codes
zipcount = merged_2021["Zip Code"].unique().shape
zipcount
```

```python
# Select only rows where Student Group ID is 1 (All students) and Grade is 13 (all grades)
filtered_2021 = merged_2021.loc[(merged_2021["Student Group ID"] == 1) & (merged_2021["Grade"] == 13), :].copy()
filtered_2021
```

```python
# Back calculate numbers of students in each category of standards met
filtered_2021["Students Standard Exceeded"] = round(filtered_2021["Percentage Standard Exceeded"]*0.01*filtered_2021["Students with Scores"])
filtered_2021["Students Standard Met"] = round(filtered_2021["Percentage Standard Met"]*0.01*filtered_2021["Students with Scores"])
filtered_2021["Students Standard Met and Above"] = round(filtered_2021["Percentage Standard Met and Above"]*0.01*filtered_2021["Students with Scores"])
filtered_2021["Students Standard Nearly Met"] = round(filtered_2021["Percentage Standard Nearly Met"]*0.01*filtered_2021["Students with Scores"])
filtered_2021["Students Standard Not Met"] = round(filtered_2021["Percentage Standard Not Met"]*0.01*filtered_2021["Students with Scores"])
filtered_2021
```

```python
# Create separate dataframes for math (Test ID 2) and english (Test ID 1)
math_2021 = filtered_2021.loc[filtered_2021["Test ID"] == 2, :].copy()
english_2021 = filtered_2021.loc[filtered_2021["Test ID"] == 1, :].copy()
```

```python
# Check to make sure math dataframe look correct
math_2021
```

```python
# Check that English dataframe looks correct
english_2021
```

```python
# Group and sum number of students math data by zip code
math_groupby_zip_2021 = math_2021.groupby("Zip Code")[[
    "Students Standard Exceeded", 
    "Students Standard Met", 
    "Students Standard Met and Above",
    "Students Standard Nearly Met",
    "Students Standard Not Met",
    "Students with Scores"
]].sum().reset_index()
math_groupby_zip_2021
```

```python
# Group and sum number of students English data by zip code
english_groupby_zip_2021 = english_2021.groupby("Zip Code")[[
    "Students Standard Exceeded", 
    "Students Standard Met", 
    "Students Standard Met and Above",
    "Students Standard Nearly Met",
    "Students Standard Not Met",
    "Students with Scores"
]].sum().reset_index()
english_groupby_zip_2021
```

```python
# Calculate percentages of students in each category for math
math_groupby_zip_2021["Percentage Standard Exceeded"] = math_groupby_zip_2021["Students Standard Exceeded"]/math_groupby_zip_2021["Students with Scores"]*100
math_groupby_zip_2021["Percentage Standard Met"] = math_groupby_zip_2021["Students Standard Met"]/math_groupby_zip_2021["Students with Scores"]*100
math_groupby_zip_2021["Percentage Standard Met and Above"] = math_groupby_zip_2021["Students Standard Met and Above"]/math_groupby_zip_2021["Students with Scores"]*100
math_groupby_zip_2021["Percentage Standard Nearly Met"] = math_groupby_zip_2021["Students Standard Nearly Met"]/math_groupby_zip_2021["Students with Scores"]*100
math_groupby_zip_2021["Percentage Standard Not Met"] = math_groupby_zip_2021["Students Standard Not Met"]/math_groupby_zip_2021["Students with Scores"]*100
math_groupby_zip_2021
```

```python
# Calculate percentages of students in each category for English
english_groupby_zip_2021["Percentage Standard Exceeded"] = english_groupby_zip_2021["Students Standard Exceeded"]/english_groupby_zip_2021["Students with Scores"]*100
english_groupby_zip_2021["Percentage Standard Met"] = english_groupby_zip_2021["Students Standard Met"]/english_groupby_zip_2021["Students with Scores"]*100
english_groupby_zip_2021["Percentage Standard Met and Above"] = english_groupby_zip_2021["Students Standard Met and Above"]/english_groupby_zip_2021["Students with Scores"]*100
english_groupby_zip_2021["Percentage Standard Nearly Met"] = english_groupby_zip_2021["Students Standard Nearly Met"]/english_groupby_zip_2021["Students with Scores"]*100
english_groupby_zip_2021["Percentage Standard Not Met"] = english_groupby_zip_2021["Students Standard Not Met"]/english_groupby_zip_2021["Students with Scores"]*100
english_groupby_zip_2021
```

```python
# Check to make sure data is reasonable size
math_groupby_zip_2021.info()
```

```python
# Check to make sure data is reasonable size
english_groupby_zip_2021.info()
```

# Pull Real Estate Data and add to each 2021 and 2022 dataframe

```python
# Read in median mortgage csv
median_mortgage = pd.read_csv("Education_Resources/median_mortgage.csv")
median_mortgage
```

```python
# do a quick check to see if the data lines up the way you think it will
plot = math_mortgage_zip_2021.plot(x="Median Mortgage", y="Percentage Standard Met and Above", kind="scatter")
```

```python
# import stats dependencies
import scipy.stats as st
```

```python
# Check pearson r value to see how strong the correlation is
median = math_mortgage_zip_2021.loc[:,"Median Mortgage"]
standard_met = math_mortgage_zip_2021.loc[:,"Percentage Standard Met and Above"]
correlation = st.pearsonr(median, standard_met)
correlation
```
