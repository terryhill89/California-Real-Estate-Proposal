## Create a merged dataset for each person's topic
## From each merged set, make one SQlite database named real_estate.db
 * Kyra = education
 * Terry = crime
 * Katejoe = demographics
 * Armann = median prices


```python
import pandas as pd
import sqlite3
```

## Education dataframe (Kyra)

```python
# Read the CSV files into Pandas DataFrames
    # English dataframe
df1e = pd.read_csv('education_Kyra/english_2021.csv')
df2e = pd.read_csv('education_Kyra/english_2022.csv')

    # Math dataframe
df1m = pd.read_csv('education_Kyra/math_2021.csv')
df2m = pd.read_csv('education_Kyra/math_2022.csv')

# Merge the DataFrames on the 'zipcode' column
merged_eng_df = df1e.merge(df2e, on='Zip Code')
merged_math_df = df1m.merge(df2m, on='Zip Code')

# Merge the two DataFrames
education_df = merged_eng_df.merge(merged_math_df, on='Zip Code')

# Convert the DataFrame to a CSV file
education_df.to_csv('education_Kyra/education_data.csv', index=False)

```

## Crime dataframe (Terry)

```python
# Convert the csv to a dataframe
crime_df = pd.read_csv('crime_Terry/merged_terry.csv')

# Convert the DataFrame to a CSV file
crime_df.to_csv('crime_data.csv', index=False)
```

## Demographics dataframe (Katejoe)

```python
# Convert the csv to a dataframe
demographics_df = pd.read_csv('demographics_Katejoe/demographics.csv')

# Convert the DataFrame to a CSV file
demographics_df.to_csv('demographics_Katejoe/demographics_data.csv', index=False)
```

## Median Prices dataframe (Armann)

```python
# Merge the county_facts and gdp_cali csv's

# csv to df
county_facts_df = pd.read_csv('median_prices_Armann/county_facts.csv')
gdp_cali_df = pd.read_csv('median_prices_Armann/GDP_Cali_2021.csv')

median_mortgage_df = pd.read_csv('median_prices_Armann/median_mortgage.csv')
zipdemos_df = pd.read_csv('median_prices_Armann/zipdemos.csv')

# Merge the DataFrames on the 'zipcode' column
merged_df1 = median_mortgage_df.merge(zipdemos_df, on='zipcode')

# Merge the DataFrames on the 'county' column
merged_df2 = county_facts_df.merge(gdp_cali_df, on='county')

# Merge the two merged dataframes into one
median_prices_df = merged_df1.merge(merged_df2, on='county')

# Convert the DataFrame to a CSV file
median_prices_df.to_csv('median_prices_Armann/median_prices_data.csv', index=False)
```

## Put all dataframes into an SQLite database

```python
# Create an SQLite connection and cursor
conn = sqlite3.connect('real_estate.sqlite')
cursor = conn.cursor()

# Write each DataFrame to a new table in the SQLite database
education_df.to_sql('education', conn, if_exists='replace', index=False)
crime_df.to_sql('crime', conn, if_exists='replace', index=False)
demographics_df.to_sql('demographics', conn, if_exists='replace', index=False)
median_prices_df.to_sql('median_prices', conn, if_exists='replace', index=False)


```
